```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Cargar el archivo CSV
df1 = pd.read_csv('anexo02.csv', encoding='ISO-8859-1')


# Eliminar filas con valores faltantes (si es necesario)
df1 = df1.dropna()

# Definir variables predictoras (X) y la variable objetivo (y)
X = df1.drop(columns=["Calidad de la carne"])
y = df1["Calidad de la carne"]

# Identificar columnas categóricas
categorical_columns = X.select_dtypes(include=['object']).columns

# Convertir columnas categóricas en variables numéricas usando one-hot encoding
X = pd.get_dummies(X, columns=categorical_columns)

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo de regresión lineal
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predecir con el modelo entrenado
y_pred_lr = lr_model.predict(X_test)

# Evaluar el modelo
print("Precisión de la Regresión Lineal:", lr_model.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Evaluar el rendimiento del modelo de regresión lineal
mse = mean_squared_error(y_test, y_pred_lr)
mae = mean_absolute_error(y_test, y_pred_lr)
r2 = r2_score(y_test, y_pred_lr)

print("Evaluación de la Regresión Lineal")
print("Error Cuadrático Medio (MSE):", mse)
print("Error Absoluto Medio (MAE):", mae)
print("R² Score:", r2)
```

Interpretación general:

Error Cuadrático Medio (MSE)
El MSE (Mean Squared Error) es una medida común utilizada para evaluar la precisión de un modelo de regresión. En este caso tenemos:

MSE: 1093909138985026.9

Este valor es extremadamente alto y sugiere que el modelo está teniendo problemas significativos en predecir los valores de salida. Un MSE típico para un problema bien modelado suele ser mucho menor, por ejemplo, entre 0.01 y 10.000.

Error Absoluto Medio (MAE)
El MAE (Mean Absolute Error) también es una métrica importante para evaluar modelos de regresión. Aquí tenemos:

MAE: 19049552.36277987

Similar al MSE, este valor es muy alto y indica que el modelo está cometiendo errores grandes en sus predicciones.

R² Score
El R² score (coefficient of determination) mide la proporción del varianza total explicada por el modelo. Tenemos:

R² Score: -571165618067095.2

Este valor negativo es anormal y sugiere que el modelo no está explicando nada de la varianza total. De hecho, parece que el modelo podría estar peor que un modelo que siempre predice el mismo valor constante.
